Data Preprocessing

```python
# drop correlated column
data=data.drop(columns=[ 'maxtemp', 'temparature', 'mintemp'])
```

```python
 data.head()
```

```python
print(data['rainfall'].value_counts())
```

```python
# seprate majority and minority class
df_majority=data[data['rainfall']==1]
df_minority=data[data['rainfall']==0]
```

```python
print(df_majority.shape)
print(df_minority.shape)
```

```python
# dwonsample majority class to match minority count

df_majority_downsampled=resample(df_majority,replace=False,n_samples=len(df_minority),random_state=422)
```

```python
df_majority_downsampled.shape
```

```python
df_dwonsampled=pd.concat([df_majority_downsampled,df_minority])
```

```python
df_dwonsampled.shape
```

```python
df_dwonsampled.head()
```

```python
# shuffle final dataset

df_dwonsampled=df_dwonsampled.sample (frac=1,random_state=42).reset_index(drop=True)
```

```python
df_dwonsampled.head()
```

```python
df_dwonsampled['rainfall'].value_counts()
```

```python
# splitting the data into train data and test data
X=df_dwonsampled.drop(columns=['rainfall'])
Y=df_dwonsampled['rainfall']
```

```python
print(X)
```

```python
print(Y)
```

```python
 #spliting the data training data and test data
 X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2,random_state=42)
```

Model Training

```python
rf_model=RandomForestClassifier(random_state=42)

param_grid_rf={
    'n_estimators':[50,100,200],
    "max_features":['sqrt','log2'],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4]
}

```

```python
# hypertuning using GrifSearchCV
grid_search_rf=GridSearchCV(estimator=rf_model,param_grid=param_grid_rf,cv=5,n_jobs=-1,verbose=2)

grid_search_rf.fit(X_train,Y_train)
```

```python
best_rf_model=grid_search_rf.best_estimator_
print("Best parameter for Random Forest :",grid_search_rf.best_params_)
```

Model Evaluation

```python
cv_score=cross_val_score(best_rf_model,X_train,Y_train,cv=5)
print("Cross validation score for Random Forest:",cv_score)
print("Mean cross validation score for Random Forest:",np.mean(cv_score))
```

```python
# test  set performance
Y_pred=best_rf_model.predict(X_test)
print("Test set Accuracy score for Random Forest:",accuracy_score(Y_test,Y_pred))
print("Confusion matrix for Random Forest:\n",confusion_matrix(Y_test,Y_pred))
print("Classification report for Random Forest:\n",classification_report(Y_test,Y_pred))
```
